Load results. 

```julia
loss_training_mlp = Matrix(CSV.read("data/AR2/loss_vec_training_mlp_1.csv"; allowmissing=:auto))
loss_val_mlp = Matrix(CSV.read("data/AR2/loss_vec_val_mlp_1.csv"; allowmissing=:auto))
predictions_mlp = Matrix(CSV.read("data/AR2/predictions_mlp_1.csv"; allowmissing=:auto))

posterior_simple_DNN = Matrix(CSV.read("data/AR2/mlp_1_abcrs_post.csv"; allowmissing=:auto))'
h1_DNN = kde(posterior_simple_DNN[1,:])
h2_DNN = kde(posterior_simple_DNN[2,:]);
println(mean(posterior_simple_DNN, dims = 2))
println(std(posterior_simple_DNN, dims = 2))
```

```julia
# plot training and val error
PyPlot.figure(figsize=(5,4))
PyPlot.plot(1:length(loss_training_mlp), loss_training_mlp[:], "*-r")
PyPlot.plot(1:length(loss_training_mlp), loss_val_mlp[:], "*-b");
```

```julia
# plot predictions
PyPlot.figure(figsize=(5,4))
PyPlot.subplot(121)
PyPlot.plt[:hist2d](y_test[:,1], predictions_mlp[:,1], bins=(100, 100))
PyPlot.subplot(122)
PyPlot.plt[:hist2d](y_test[:,2], predictions_mlp[:,2], bins=(100, 100));
```
